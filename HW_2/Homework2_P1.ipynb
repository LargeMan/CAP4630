```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

f1 = lambda x, y: np.cos(-(x + y))
f2 = lambda x, y: 0.5 * np.sin(x**2 + y**2 + x * y)
f3 = lambda x, y: (x**3) + (y**3)

```

```
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)

X, Y = np.meshgrid(x, y)

Z1 = f1(X, Y)
Z2 = f2(X, Y)
Z3 = f3(X, Y)
```

#Minimum of F1

```
fig = plt.figure()
ax = Axes3D(fig)
ax.contour3D(X, Y, Z1, 100, cmap='Greens');
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

p1, _ = ax.get_zlim()
i = np.where(Z1 == p1)[0][0]
j = np.where(Z1[i] == p1)[0][0]

ax.plot([X[i,j]], [Y[i,j]], [Z1[i,j]], 'ro')

print("Min of {}".format(p1))
```

```
plt.contour(X, Y, Z1, colors='black');
```

```
plt.contourf(X, Y, Z1, 20, cmap='RdGy')
plt.colorbar();
```

#Maximum of F2

```
# test
fig = plt.figure()
ax = Axes3D(fig)

ax.contour3D(X, Y, Z2, 100, cmap='binary');
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#p2 = np.amax(Z2)
_, p2 = ax.get_zlim()

i = np.where(Z2 == p2)[0][2]
j = np.where(Z2[i] == p2)[0][0]
print("Max at Z = {}.".format(p2))

ax.plot([X[i,j]], [Y[i,j]], [Z2[i,j]], 'ro', markersize = 5)
```

```
plt.contour(X, Y, Z2, colors='black');
```

```
plt.contourf(X, Y, Z2, 20, cmap='RdGy')
plt.colorbar();
```
