# Import MNIST dataset

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
from keras.datasets import mnist

init, final = mnist.load_data()
```

# Find the averages of each digit image, and display

```
def count_total(n, data):
  X, Y = data
  count = 0
  total = np.zeros(shape=(28,28))

  # Adds all images in X with indices of Y where Y == n
  for i in [j for j in range(len(Y)) if Y[j] == n]:
    count += 1
    total += X[i]

  return count, total
  
for i in range(10):
  # Find count and total, then calculate average
  cnt1, tot1 = count_total(i, init)
  cnt2, tot2 = count_total(i, final)
  average = (tot1 + tot2) / (cnt1 + cnt2)
  
  plt.figure(i+1)
  plt.imshow(average, cmap = 'twilight')
  plt.title('Digit {}'.format(i))
```
